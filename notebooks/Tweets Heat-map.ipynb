```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from datetime import datetime
import calmap
```

```python
df = pd.read_csv(r"C:\Users\Usuario\Downloads\geo_twitter_data.csv")
datos = pd.read_csv(r"C:\Users\Usuario\Downloads\geo_twitter_data_names.csv")

datoslib=pd.read_csv(r"C:\Users\Usuario\Downloads\geo_twitter_data_por.csv")
df
```

```python
nombre = ((np.asarray(datos['nombre'])))
fig, ax = plt.subplots(figsize=(9,7))
sns.heatmap(df,ax=ax, yticklabels=nombre, cbar_kws={'label': 'Covid-19 Tweets'} )
ax.set_title('Covid-19 Tweets in Mexico by state')
plt.show()
```
